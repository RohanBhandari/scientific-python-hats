```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
```

```python
df = pd.read_pickle('new_ak5_withRF.pkl')
```

```python
print df.columns
df.head()
```

```python
plot_vars=['jet_pt','jet_eta','jet_electron_frac','jet_muon_frac','jet_photon_frac','jet_neutHad_frac','jet_charHad_frac','jet_electron_mult','jet_muon_mult','jet_photon_mult','jet_neutHad_mult','jet_charHad_mult']
for v in plot_vars : 
    plt.hist(df[v],bins=100)
    plt.xlabel(v)
    #plt.yscale('log')
    plt.show()
```

```python
def jes_profile(indep_var,dep_var,c,low_bin,high_bin,nbins):
    print "plot_var:",indep_var
    print "low_bin:",low_bin
    print "high_bin:",high_bin
    print "nbins:",nbins
    df[indep_var+'_bins'] = pd.cut(df[indep_var],np.arange(low_bin,high_bin+0.00001,(high_bin-low_bin)/float(nbins)),labels=range(nbins))
    means = df.groupby([indep_var+'_bins']).mean()[dep_var].values
    counts = df.groupby([indep_var+'_bins']).count()[dep_var].values
    sqrt_counts = map(np.sqrt,counts)
    errs  = df.groupby([indep_var+'_bins']).std()[dep_var].values
    errs = errs/sqrt_counts
    bin_center=np.arange(low_bin,high_bin+0.00001,(high_bin-low_bin)/float(nbins))
    bin_center = bin_center[:-1]
    bin_center = map(lambda x : x+(high_bin-low_bin)/float(nbins)/2.,bin_center)
    plt.errorbar(x=bin_center,y=means,yerr=errs,fmt='o',color=c)
    plt.xlabel(indep_var)
    plt.ylabel("<residual>")
    plt.ylim(0.8,1.1)
```

```python
var_binning = [(0,1000,90),(-3,3,60),(0,1.5,50),(0,1.5,50),(0,1.5,50),(0,1.5,50),(0,1.5,50),(0,100,100),(0,100,100),(0,100,100),(0,100,100),(0,100,100)]
for f,b in zip(plot_vars,var_binning) : 
    plt.hist2d(df[f],df['jet_jes'],weights=df['mcweight'],bins=100,norm=LogNorm())
    plt.ylabel('jet_jes')
    plt.grid()
    plt.colorbar()
    plt.xlabel(f)
    plt.show()
```

```python
for f,b in zip(plot_vars,var_binning) : 
    jes_profile(f,'jet_jes','g',*b)
    plt.ylim(0.,1.1)
    plt.show()
```

```python
df['jet_px'] = map(np.cos,df['jet_phi'])*df['jet_pt']
df['jet_py'] = map(np.sin,df['jet_phi'])*df['jet_pt']
df['jet_px_truth'] = map(np.cos,df['jet_phi'])*df['jet_pt']*df['jet_jes']
df['jet_py_truth'] = map(np.sin,df['jet_phi'])*df['jet_pt']*df['jet_jes']
df['jet_pt_truth'] = df['jet_pt']*df['jet_jes']

df['jet_px_rfall'] = map(np.cos,df['jet_phi'])*df['jet_pt']*df['prediction_rfall']
df['jet_py_rfall'] = map(np.sin,df['jet_phi'])*df['jet_pt']*df['prediction_rfall']
df['jet_pt_rfall'] = df['jet_pt']*df['prediction_rfall']

df['jet_px_rfpteta'] = map(np.cos,df['jet_phi'])*df['jet_pt']*df['prediction_rfpteta']
df['jet_py_rfpteta'] = map(np.sin,df['jet_phi'])*df['jet_pt']*df['prediction_rfpteta']
df['jet_pt_rfpteta'] = df['jet_pt']*df['prediction_rfpteta']

df_evt = df.groupby(['run','lumi','event']).sum()
df_evt['mht'] = map(np.sqrt,np.multiply(df_evt['jet_px'],df_evt['jet_px'])+np.multiply(df_evt['jet_py'],df_evt['jet_py']))
df_evt['mht_truth'] = map(np.sqrt,np.multiply(df_evt['jet_px_truth'],df_evt['jet_px_truth'])+np.multiply(df_evt['jet_py_truth'],df_evt['jet_py_truth']))
df_evt['mht_rfall'] = map(np.sqrt,np.multiply(df_evt['jet_px_rfall'],df_evt['jet_px_rfall'])+np.multiply(df_evt['jet_py_rfall'],df_evt['jet_py_rfall']))
df_evt['mht_rfpteta'] = map(np.sqrt,np.multiply(df_evt['jet_px_rfpteta'],df_evt['jet_px_rfpteta'])+np.multiply(df_evt['jet_py_rfpteta'],df_evt['jet_py_rfpteta']))
df_evt.head()
```
